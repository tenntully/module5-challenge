### Background (adapted from Boot Camp Module 5 Challenge)
Pymaceuticals, Inc., is a new pharmaceutical company that specializes in anti-cancer medications. Recently, it began screening for potential treatments for squamous cell carcinoma (SCC), a commonly occurring form of skin cancer.

In their most recent animal study, 249 mice who were identified with SCC tumors received treatment with a range of drug regimens. Over the course of 45 days, tumor development was observed and measured. The purpose of this study was to compare the performance of Pymaceuticalsâ€™ drug of interest, Capomulin, against the other treatment regimens.


### Analysis
My findings for the Pymaceutials challenge are summarized as follows:
-1893 total data points for 249 mice.
-Mean, Median, Variance, Std Dev indicate that, for each drug regimen, the tumor sizes had a centralized normal distribution of data.  
-On average, over the period of treatment, the overall tumor size was lower for Capomulin and Ramicane compared to other drug regimens based on mean data.
-The total populations of mice tesetd for Capomulin and Ramicane was higher
-Even number of male and female mice.  With enough data, there should not be a sex bias in the results. 
-Box plots reveal that the IQR for final tumor volume is tighter for Capomulin and Ramicane than for Infubinol and Ceftamin.  Also the entire data distribution for final tumor volume was much lower for Capomulin and Ramicane (generally 31 mm^3 to 40 mm^3) than for Infubinol and Ceftamin (generally 50 mm^3 to 65 mm^3).  This appears to indicate postive effectiveness in reducing tumor size for Capomulin that is similar to Ramicane treatment.  
-Line plots for mouse ID l509 appears to indicate reductions in tumor size starting 20 days after treatment and generally shrinking for the next 40 days.  More investigation should be done by plotting many mice for each of the treatments and look at general trends. 
-Plotting average tumor size versus mouse weight shows that the there appears to be a good correlation between the two.  The Pearson correlation between mouse weight and average tumor volume for the Capomulin regimen is 0.84 which is relatively close to 1.  Larger mice generally have larger average tumor sizes over the period of treatment.  
 

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#Combine data into a single file; combine on Mouse ID
mouse_test_data = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
# len(mouse_test_data) : 1893 total data points
mouse_test_data.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

tumor_mean_df = mouse_test_data_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
tumor_median_df = mouse_test_data_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
tumor_variance_df = mouse_test_data_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
tumor_stddev_df = mouse_test_data_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
tumor_SEM_df = mouse_test_data_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
column_names = ["Mean Tumor Volume", "Median Tumor Volume", "Tumor Volume Variance", "Tumor Volume Std Dev", "Tumor Volume Std Error"]

# Produce DataFrame of values with column headers
tumorvolume_summary_statsDF = pd.concat([tumor_mean_df,tumor_median_df, tumor_variance_df, 
                                         tumor_stddev_df, tumor_SEM_df],axis=1)
tumorvolume_summary_statsDF.columns = column_names

tumorvolume_summary_statsDF
```

```python
# Using the aggregation method, produce the same summary statistics in a single line.
tumorvolume_summary_statsDF_alt = mouse_test_data_clean.groupby('Drug Regimen').agg(
    {'Tumor Volume (mm3)':['mean', 'median', 'var', 'std', 'sem']})

tumorvolume_summary_statsDF_alt
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# loc on the four regimens only
mouse_test_data_4regimens = mouse_test_data_clean.loc[mouse_test_data_clean['Drug Regimen'].isin(
    ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]
mouse_test_data_4regimens = mouse_test_data_4regimens.sort_values(by=['Mouse ID', "Timepoint"])

# Get the last (greatest) timepoint for each mouse
per_mouse_finaltumorvol_df = mouse_test_data_4regimens.groupby('Mouse ID').tail(1)
per_mouse_finaltumorvol_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)

# Create a list of drug regimens
drug_list = per_mouse_finaltumorvol_df['Drug Regimen'].unique()

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data_df = pd.DataFrame(columns = ["Drug Regimen", "Lower Bound", "Upper Bound"])

# Calculate quantiles, IQR, upper and lower bounds for each drug regimen
for drug in drug_list:
    finaltumorvol = per_mouse_finaltumorvol_df.loc[per_mouse_finaltumorvol_df['Drug Regimen'] == drug,
            ['Tumor Volume (mm3)']]

    Q1 = np.quantile(finaltumorvol, 0.25)
    Q3 = np.quantile(finaltumorvol, 0.75)
    iqr = Q3-Q1
    
    lower_bound = Q1 - (1.5*iqr)
    upper_bound = Q3 + (1.5*iqr)
    
    # Create a list of lower and upper bound for active drug regemin and append list to the overall DF
    tumor_vol_data_list = (drug, lower_bound, upper_bound)
    tumor_vol_data_df.loc[len(tumor_vol_data_df)] = tumor_vol_data_list

# Create a new final tumor volume DF that includes lower and upper bound     
drug_effectivess_df = pd.merge(
    per_mouse_finaltumorvol_df, tumor_vol_data_df, how="left", on=["Drug Regimen", "Drug Regimen"])

# Create a boolean identifier for each row that is inside or outside of the bounds
drug_effectivess_df["Potential Outlier"] = (
    drug_effectivess_df['Tumor Volume (mm3)'] > drug_effectivess_df['Upper Bound']) | (
    drug_effectivess_df['Tumor Volume (mm3)'] < drug_effectivess_df['Lower Bound'])

# Create a DF of potential outliers
drug_effectiveness_outliers = drug_effectivess_df.loc[drug_effectivess_df['Potential Outlier']]
drug_effectiveness_outliers
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group. 
# Capomulin, Ramicane, Infubinol, and Ceftamin
Capomulin_data = per_mouse_finaltumorvol_df.loc[per_mouse_finaltumorvol_df['Drug Regimen'] == 'Capomulin']['Tumor Volume (mm3)']
Ramicane_data = per_mouse_finaltumorvol_df.loc[per_mouse_finaltumorvol_df['Drug Regimen'] == 'Ramicane']['Tumor Volume (mm3)']
Infubinol_data = per_mouse_finaltumorvol_df.loc[per_mouse_finaltumorvol_df['Drug Regimen'] == 'Infubinol']['Tumor Volume (mm3)']
Ceftamin_data = per_mouse_finaltumorvol_df.loc[per_mouse_finaltumorvol_df['Drug Regimen'] == 'Ceftamin']['Tumor Volume (mm3)']

fig, ax = plt.subplots(figsize=(6, 4))

# Remove y-axis tick marks and add major gridlines for y-axis
ax.yaxis.set_ticks_position('none')
ax.grid(color='grey', axis='y', linestyle='-', linewidth=0.25, alpha=0.5)

# Set plot title and y-axis label
ax.set_title('Distribution of Tumor Volume for Each Treatment Group')
ax.set_ylabel('Final Tumor Volume (mm3)')

# Set data and labels
dataset = [Capomulin_data, Ramicane_data, Infubinol_data, Ceftamin_data]
labels = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Generate boxplot
ax.boxplot(dataset, labels=labels)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Index Mouse ID and DrugRegimen
MouseID = 'l509'
DrugRegimen = 'Capomulin'

# Create new DF of data filtered by the Drug Regimen and Mouse ID
Capumulin_over_time_df = mouse_test_data_clean.loc[(mouse_test_data_clean['Drug Regimen'] == DrugRegimen) & 
                    (mouse_test_data_clean['Mouse ID'] == MouseID)]

# Generate line chart
plt.plot(Capumulin_over_time_df['Timepoint'], Capumulin_over_time_df['Tumor Volume (mm3)'], marker="o", color="red", linewidth=1)
plt.title(f'{DrugRegimen} Treatment of Mouse {MouseID} Over Time')
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# loc on Capolumin regimen only
mouse_test_data_Capomulin = mouse_test_data_clean.loc[mouse_test_data_clean['Drug Regimen'] == 'Capomulin']

# Calculate average tumor volume and mouse weight for each mouse and put in a new DF
per_mouse_weight = mouse_test_data_Capomulin.groupby(["Mouse ID"]).mean()["Weight (g)"]
per_mouse_tumor = mouse_test_data_Capomulin.groupby(["Mouse ID"]).mean()["Tumor Volume (mm3)"]
per_mouse_df = pd.concat([per_mouse_weight, per_mouse_tumor], axis=1)

# Generate scatter plot with new DF
plt.scatter(per_mouse_df['Weight (g)'], per_mouse_df['Tumor Volume (mm3)'], marker="o", color="red", linewidth=1)
plt.title('Mouse Tumor Volume vs Weight')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Calculate Pearsons correlation coefficient
correlation = st.pearsonr(per_mouse_weight, per_mouse_tumor)

# Calculate linear regression attributes
(slope, intercept, rvalue, pvalue, stderr) = linregress(per_mouse_weight, per_mouse_tumor)
regress_values = per_mouse_weight * slope + intercept

# Generate scatter plot with linear regression
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(per_mouse_weight, per_mouse_tumor)
plt.plot(per_mouse_weight,regress_values,"r-")
plt.annotate(line_eq,(20,36),fontsize=10,color="red")
plt.title('Mouse Tumor Volume vs Weight')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
print(f"The correlation between mouse weight and average tumor volume for the Capomulin regimen is {round(correlation[0],2)}")
plt.show()
```
